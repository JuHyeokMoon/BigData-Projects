```
data = pd.read_csv("N_cont&like.csv")
```

```
# 댓글 데이터 한 변수로 연결
N_data = data['Content'][0]
for i in range(1, len(data)):
    N_data = N_data + data['Content'][i]
N_data
```

```
import nltk
nltk.download('punkt')
```

```
# 토큰화
from nltk.tokenize import word_tokenize
tag = word_tokenize(N_data)
```

```
# 토큰화
from nltk.tokenize import RegexpTokenizer
# [\w]+ 문자 또는 숫자가 1개 이상인 경우를 인식
retokenize = RegexpTokenizer("[\w]+")
retokenize.tokenize(N_data)
```
